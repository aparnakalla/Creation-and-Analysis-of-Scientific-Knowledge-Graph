```python
topic_no=[]
for topics in venues['Name']:
    if user_input in topics:
        topic_no.append(topics.split('_')[0])
#print(topic_no)

Titles=[]
Scores=[]
for i in range(len(topic_no)):
    Titles.extend(final_score[final_score['Topic_NO']==int(topic_no[i])]['Title'].tolist())
    Scores.extend(final_score[final_score['Topic_NO']==int(topic_no[i])]['final_score'].tolist())

Papers={}
for i in range(len(Titles)):
    Papers[Titles[i]]=Scores[i]

    
Papers=dict(sorted(Papers.items(), key=lambda item:item[1],reverse=True))
result = dict(itertools.islice(Papers.items(), 10)) 
for i,j in zip(result,result.values()):
    print(i,j)
```

```python
topics=[]
topicn=[]
for i in venues['Name']:
    if(user_input in i):
        xx=i.index('_')
        topic_no=i[:xx]
        topics.append(topic_no)
        topicn.append(i[xx+1:])
        print(i[xx+1:],topic_no)

for i in range(len(topics)):     
    fig, ax = plt.subplots(figsize=(8, 8))
    sns.lineplot(data=df[int(topics[i])],x='YOP',y='Count').set(title='Count of papers in the topic '+ user_input+" (keywords include " +topicn[i]+')')


import matplotlib.pyplot as plt
fig = plt.figure()
count=[]

count1=[0]*len(topics)
count2=[0]*len(topics)

sim_topics=[]
for i in data:
    if(i==user_input):
        for j in range(1,5):
            if(data[user_input][1][j]>0.3):
                sim_topics.append(data[user_input][0][j]-1)
                print('similar items to '+user_input+':', venues['Name'][data[user_input][0][j]])
            
        #print(data[inputt][1])
for i in range(len(sim_topics)):     
    fig, ax = plt.subplots(figsize=(8, 8))
    sns.lineplot(data=df[int(sim_topics[i])],x='YOP',y='Count').set(title='Count of papers in the topic related to '+ user_input+ "(keywords include"+ venues['Name'][sim_topics[i]]+")")
```
